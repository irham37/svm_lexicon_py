```python
from joblib import load
import ngram as ng
import pandas as pd
```

```python
import os
import pymysql
import pymysql.cursors
```

```python
host = os.getenv('localhost')
port = os.getenv('3306')
user = os.getenv('root')
password = os.getenv('')
database = os.getenv('simsmansa')

conn = pymysql.connect(
    host=host,
    port=int(3306),
    user="root",
    passwd=password,
    db="simsmansa",
    charset='utf8mb4')
cursor=conn.cursor()
```

```python
# load the model from disk TF-idf
filename = 'model/tfdf_model_lexicon.joblib'
tfidf = load(filename)

# load the model from disk SVM
filename = 'model/svm_model_lexicon.joblib'
SVM = load(filename)
```

```python
proses = "lebih baik lagi"
proses = ng.ngram(proses)
tfidf_matrix_ng = tfidf.transform([proses])

prediksi_lx = SVM.predict(tfidf_matrix_ng.toarray())[0]
print("  "+proses)
print("<< N-Gram+lexicon  : "+prediksi_lx+">>")
```

```python
def prediksi(proses):
    proses = ng.ngram(proses)
    tfidf_matrix_ng = tfidf.transform([proses])
    prediksi_lx = SVM.predict(tfidf_matrix_ng.toarray())[0]
    return prediksi_lx

prediksi(proses)
```

```python
df = pd.read_sql_query("SELECT a.*,b.catatan_kuesioner as komentar FROM komentar_siswa_data_bersih a,siswa_mengajar b where a.id=b.id_siswa_mengajar;",conn)
df.head(10)
n_data = list()
id_ = df['id'].tolist()
asli = df['komentar'].tolist() #catatan_kuesioner harus diganti dengan data kotor/data belum masuk preprocessing
for i in df['komentar'].tolist():
    n_data.append(i.replace("[","").replace("]","").replace('"',"").replace(','," "))
n_data
```

```python
pred_list = list()
for p, ix in zip(n_data, id_):
    hasil = prediksi(p)
    pred_list.append(hasil)
#     print(hasil)
    sql = "UPDATE komentar_siswa_data_bersih SET prediksi_svm_lexicon = '%s' WHERE `id` = '%s'" % (hasil, ix)
    print(hasil, ix)
    try:
        cursor.execute(sql)
        conn.commit()
    except:
        conn.rollback()
```

```python
dictt={
    'id':id_,
    "label prediksi":pred_list,
    "komentar":n_data,
    "komentar_asli":asli,
}
df=pd.DataFrame.from_dict(dictt)
df.to_excel("hasil_prediksi.xlsx")
```
